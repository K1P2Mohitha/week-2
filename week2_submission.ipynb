```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split

# Example DataFrame
df = pd.DataFrame({
    'FileName': ['img1.jpg', 'img2.jpg', 'img3.jpg', 'img4.jpg', 'img5.jpg', 'img6.jpg'],
    'FilePath': ['dataset/trash/img1.jpg', 'dataset/paper/img2.jpg', 'dataset/plastic/img3.jpg', 'dataset/cardboard/img4.jpg', 'dataset/metal/img5.jpg', 'dataset/glass/img6.jpg']
})

# Add label column
df['label'] = df['FilePath'].apply(lambda x: os.path.basename(os.path.dirname(x)))

print(df.head())

# Split
X = df.drop('label', axis=1)
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```
